```python
算子：
Linear(全连接层)--作用：对输入做线性变换：y=xW^T+b
                --数学本质：进行矩阵乘法+偏置
                --适用于特征映射、分类输出层
Cpnv（卷积层）--作用：用卷积核提取局部空间特征
              --数学本质：二维/三维互相关运算
              --适用于图像/序列特征提取（CNN核心）
MN（最大池化/平均池化）--作用：下采样降维，保留关键特征
                       --数学本质：局部取最大/平均值
                       --适用于减少计算量、防止过拟合
LSTM（长短期记忆网络）--作用：时序数据建模，解决长序列梯度消失
                      --数学本质：门控循环单元
                      --适用于文本、语音、时间序列任务
```

```python
损失函数（做不同任务，实际上就是损失函数不同）--当我们做完结果后才能开始评估的
定义：损失函数衡量当前本组权重参数做完之后的一个结果
（衡量模型预测值与真实标签的差异，为优化算法提供梯度方向，是训练的核心目标）
ps：损失函数的值相同时，两个模型不一定一致（权重参数变化可能不同）
损失函数=数据损失＋正则化惩罚项
1.交叉熵：适用于多分类任务（Softmax分类器）
2.KL散度
3.均方误差：适用于回归任务、连续值预测
4.平方绝对误差
```

```python
正则化的作用：（正则惩罚力度对结果的影响）
1.正则惩罚力度比较小的情况下：整体的模型不太合适，模型复杂度高（容易过拟合，效果不太好）
2.正则惩罚力度比较大的情况下：模型被过度约束，且过于简单，容易出现欠拟合，学习能力不足
3.正则惩罚力度适中的情况下：力度适中，模型复杂度合理
-补：正则化的本质：在损失函数中加入惩罚项，约束模型参数复杂度，防止过拟合
```

```python
回归任务：由得分值去计算一个损失
分类任务：由概率值去计算一个损失

```

```python
神经网络：
1.前向传播
2.反向传播（可以更新模型（梯度下降））
```

```python
反向传播：（几种常见的门单元（当做一种操作））
1.加法门单元：均等分配--做加法x+y=q（梯度均匀地传给x和y）
2.MAX门单元：给最大的 max(0,x)--做计划函数会用到的（把梯度传给最大的，小的那个则没有）
3.乘法门单元：互换的感觉--做乘法xy=q
```

```python
链式法则：梯度是一步一步进行传播的（从后往前一步一步去做，所以是反向传播），不能直接进行最后一步。
```

```python
激活函数：
--为网络引入非线性能力，让网络可以学习复杂模式，核心作用时打破线性变换的限制
1.Sigmpid--输出(0,1),可做概率；梯度易消失（由于个别层的影响）；
--用于二分类输出层，早期MLP
2.ReLU--计算简单，缓解梯度消失；负区间梯度为0（比较实用且常用）（不会出现梯度消失的现象）
--用于CNN、MLP隐藏层（最常用）
3.GELU--平滑非线性，自适应门控
--用于Transformer、大模型隐藏层
4.SwiGLU--结合Swish与GLU，提升大模型性能
--LLM、Transformer前馈层
```
